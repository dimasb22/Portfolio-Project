```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import make_blobs   # "make_blobs" function used to build the toy dataset
from sklearn.cluster import KMeans
```

```python
# retrieve dataset with function "make_blobs"

X, y_true = make_blobs(n_samples=300,   # mengambil dataset berisi 300 sampel/baris
                       n_features=2,    # mengambil dataset berisi 2 kolom
                       centers=4,       # akan membuat data menjadi 4 kelompok/klaster
                       cluster_std=0.60, # akan terlihat jika diplot,  cluster_std(standar deviasi) fungsinya untuk mengubah sebaran dari data
                       random_state=0)   # akan terlihat jika diplot, berfungsi jika kita ingin menyeragamkan data agar saat di run tidak berubah terus, nilainya yg dimasukkan terserah

plt.scatter(X[:, 0], X[:, 1], s=50);    # "(X[:, 0]" ini sumbu x nya.  "X[:, 1]" sumbu y nya, s=50 adalah size
```

```python
# Building k-means

kmeans = KMeans(n_clusters=4)   # n_cluster harus sama dengan jumlah centers yg dibuat diatas
kmeans.fit(X)                   # ini menyuruh k means untuk belajar dari data x
y_kmeans = kmeans.predict(X)    # untuk mengompokkan data2 yg ada di x dan dimasukkan dalam var "y_kmeans"

# Data plotting process
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis') # c=y_kmeans, utk menambah parameter warna

# Determine the location of the centroid and plotting
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
