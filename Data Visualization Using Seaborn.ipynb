# Import Library

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Data

```python
# we use dataset from seaborn

dataset = sns.load_dataset('mpg')
```

```python
dataset.head()
```

# Making Histogram

```python
sns.histplot(data=dataset, x ='mpg', bins = 30)
```

## Making boxplot

```python
sns.boxplot(data = dataset, x='mpg')
```

## Making Bar Chart

```python
sns.barplot (data=dataset, x='origin', y='mpg')
```

## making lineplot

```python
sns.lineplot (data=dataset, x = 'model_year', y= 'mpg', hue = 'origin')
```

# Making Saterplot

```python
sns.scatterplot(data=dataset, x='displacement', y='mpg')
```

# making Pie chart

```python
# to making pie chart weshould prepare the variabel and data

data_origin = dataset['origin'].value_counts().to_dict()
data_pie = data_origin.values()
label_pie = data_origin.keys()
colors = sns.color_palette('pastel')[0:5]
```

```python
plt.pie(x = data_pie, labels = label_pie, colors = colors, autopct ='%.0f%%')
plt.title("Origin")
plt.show()
```
